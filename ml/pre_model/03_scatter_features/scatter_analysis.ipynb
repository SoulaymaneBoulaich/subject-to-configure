### Reading the plots
| Pattern | Meaning |
|---|---|
| Points close to the line | Strong linear relationship |
| Wide cloud | Weak or noisy relationship |
| Curved pattern | Non-linear relationship → may need polynomial features |
| Funnel shape | Heteroscedasticity — variance grows with x |

## Engineered Features

Beyond the raw features, five interaction/ratio terms were created:

| Feature | Formula | Rationale |
|---|---|---|
| `total_screen` | screen + tech + work hours | Total daily screen exposure |
| `leisure_screen` | social_media + gaming hours | Passive/recreational screen time |
| `activity_ratio` | physical_activity / (screen + 1) | Balance between movement and sedentary behaviour |
| `sleep_stress_interact` | sleep_hours × (10 − stress_level) | Restorative sleep under low stress |
| `support_online` | support_systems × online_support | Combined support network |

The interaction term `sleep_stress_interact` achieved r = +0.090 with the target — the strongest among engineered features.
