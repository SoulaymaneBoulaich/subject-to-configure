## 1. 10-Fold Cross-Validation

**Protocol:** K=10, shuffled, random_state=42

## 2. Regularization Paths
`fig07_regularization_paths.png`

### Ridge (L2) — left panel
$$\hat{\beta}_{Ridge} = (X^TX + \alpha I)^{-1} X^T y$$

As $\alpha$ increases (left to right on the x-axis):
- All coefficients shrink **smoothly toward 0**
- No feature is ever exactly zeroed
- The vertical dashed line marks the best $\alpha$ chosen by RidgeCV

### Lasso (L1) — right panel
$$L_{Lasso} = \|y - X\beta\|^2 + \alpha \|\beta\|_1$$

As $\alpha$ increases:
- Coefficients hit **exactly 0** (lines touch the axis) — feature selection!
- The order lines reach 0 reveals feature importance
- Best $\alpha$ chosen by LassoCV (10-fold)

## 3. Best-Fit Line — Predicted vs Actual
`fig08_best_fit_line.png`

The diagonal dashed line = **perfect prediction** (ŷ = y).  
Points close to this line = accurate model.

- A **fan shape** → heteroscedasticity (errors grow with y)
- A **banana curve** → missing non-linearity
- **Vertical stripes** → the target has discrete clusters (as seen here around y=14.8)

## 4. Residual Diagnostics
`fig09_residual_diagnostics.png`

Residuals $\varepsilon = y - \hat{y}$ should satisfy the Gauss-Markov conditions:

| Panel | Test | Ideal |
|---|---|---|
| Residuals vs Fitted | Homoscedasticity | Random cloud around 0 |
| Q-Q Plot | Normality of errors | Points on diagonal |
| Histogram | Normality | Bell-shaped, centred at 0 |
| Scale-Location | Homoscedasticity | Flat trend line |
| Residual Series | No autocorrelation | No pattern over time |
| ACF | No autocorrelation | Bars inside blue band |

**Durbin-Watson statistic = 1.99** (ideal ≈ 2) → no autocorrelation ✓  
**Breusch-Pagan proxy p < 0.05** → heteroscedasticity detected → consider WLS or log-transform of target.
