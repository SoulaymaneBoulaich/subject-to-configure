# Exploratory Data Analysis (EDA)

This notebook explains every artefact produced during the EDA phase of the Pre-ML study.

## 1. Correlation Table
`correlation_table.csv` — Pearson **r** and Spearman **ρ** of every feature with the chosen target `mental_wellness_index`.

| Column | Meaning |
|---|---|
| `pearson_r` | Linear correlation (sensitive to outliers) |
| `pearson_p` | p-value — is the linear correlation statistically significant? |
| `spearman_r` | Rank-based correlation (robust to outliers & non-linearity) |
| `spearman_p` | p-value for the rank correlation |

**Rule of thumb:** |r| > 0.3 = moderate, |r| > 0.5 = strong.

```python
import pandas as pd
corr = pd.read_csv('correlation_table.csv')
corr.sort_values('pearson_r', key=abs, ascending=False)
```

## 2. Target Distribution
`fig01_target_distribution.png`

- **Left panel** — raw histogram: the target is *extremely* right-skewed (skewness ≈ 17.5). Most values cluster around 14.8.
- **Centre panel** — Q-Q plot: points deviate sharply from the diagonal → the target is **not normally distributed**.
- **Right panel** — `log1p` transform flattens the spike somewhat. A power transform (Box-Cox / log) may be needed before modelling.

**Normality tests performed:**

| Test | Result |
|---|---|
| Shapiro-Wilk | PASS (artefact of large N) |
| Kolmogorov-Smirnov | FAIL — distribution ≠ Normal |
| D'Agostino K² | FAIL — excess skewness & kurtosis |

## 3. Feature Boxplots
`fig02_boxplots.png`

Boxplots reveal the spread, median, and outlier count for every numeric feature.

- **Box** = IQR (25th–75th percentile)
- **Line inside box** = median
- **Whiskers** = 1.5 × IQR
- **Dots beyond whiskers** = outliers

Features like `age`, `mood_level`, and `productivity_score` have high outlier counts (>25%). This is expected in survey data and does not necessarily mean errors.

## 4. Pearson Correlation Heatmap
`fig03_correlation_heatmap.png`

The lower-triangular heatmap shows pairwise Pearson correlations across all features + target.

- **Red** = strong positive correlation
- **Blue** = strong negative correlation
- **White** = near-zero correlation

Key observations:
- `productivity_score` & `mental_wellness_index` are positively correlated (r ≈ 0.49)
- `work_screen_hours` and `sleep_quality` show a moderate negative relationship
- The occupation dummies (`occ_*`) are collinear with each other (expected — they form one hot-encoded group)
