# Train / Validation / Test Splits

This folder contains the final prepared arrays used for model training, selection, and evaluation.

## Split Strategy

A **70 / 15 / 15** random split was used:

| Set | Purpose | Rows |
|---|---|---|
| **Train** | Fit model parameters (β) | 24,518 |
| **Validation** | Tune hyperparameters (α in Ridge/Lasso) | 5,255 |
| **Test** | Final unbiased evaluation — **never touched during training** | 5,255 |

`random_state=42` ensures full reproducibility.

## Feature Scaling

All X files are **StandardScaler-transformed**:

$$z = \frac{x - \mu_{train}}{\sigma_{train}}$$

- `\mu` and `\sigma` computed **only on X_train** → no data leakage
- The same statistics are applied to X_val and X_test
- y vectors are kept in **original scale** (not standardised)

## Distribution Check — Does Each Split Have Similar Target Stats?

```python
import pandas as pd, matplotlib.pyplot as plt

y_tr = pd.read_csv('y_train.csv')['mental_wellness_index']
y_va = pd.read_csv('y_val.csv')['mental_wellness_index']
y_te = pd.read_csv('y_test.csv')['mental_wellness_index']

print(f'Train  mean={y_tr.mean():.3f}  std={y_tr.std():.3f}')
print(f'Val    mean={y_va.mean():.3f}  std={y_va.std():.3f}')
print(f'Test   mean={y_te.mean():.3f}  std={y_te.std():.3f}')

fig, ax = plt.subplots(figsize=(10, 4))
ax.hist(y_tr, bins=60, alpha=0.5, label='Train',  color='#2ecc71')
ax.hist(y_va, bins=60, alpha=0.5, label='Val',    color='#3498db')
ax.hist(y_te, bins=60, alpha=0.5, label='Test',   color='#e74c3c')
ax.set_title('Target Distribution Across Splits')
ax.legend(); plt.tight_layout(); plt.show()
```
